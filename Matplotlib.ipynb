```python
import matplotlib.pyplot as plt
```

```python
%matplotlib inline
```

```python
import numpy as np
x = np.linspace(0,5,11)
y = x **2
```

```python
x
```

```python
y
```

```python
#functional method
plt.plot(x,y,'g-')
plt.xlabel(' X lable')
plt.ylabel('Y lable')
plt.title('TITLE')
```

```python
plt.subplot(1,2,1)
plt.plot(x,y,'r-')

plt.subplot(1,2,2)
plt.plot(y,x,'g')
 

```

```python
#object oriented method
fig = plt.figure()
axes = fig.add_axes([.1,.1,.8,.8])
axes.plot(x,y)
axes.set_xlabel('X LABLE')
axes.set_ylabel('Y LABLE')
axes.set_title('SET TITLE')
```

```python
fig = plt.figure()
axes1 = fig.add_axes([.1,.1,.8,.8])
axes2 = fig.add_axes([0.2,.5,.4,.3])
axes1.plot(x,y)
axes1.set_title('LARGER PLOT')
axes2.plot(y,x)
axes2.set_title('SMALLER PLOT')
```

```python
fig,axes = plt.subplots(nrows = 1,ncols=2)
plt.tight_layout()
## here we were not supposd to give the value for the axeses they are automatically arranged.
for current_ax in axes:
    current_ax.plot(x,y)
```

```python
fig,axes = plt.subplots(nrows = 1,ncols=2)
fig,axes[0].plot(x,y)
```

```python
fig = plt.figure(figsize=(4,4))
ax = fig.add_axes([0,0,1,1])
ax.plot(x,x**2,label='X SQUARED')
ax.plot(x,x**3,label = 'X CUBED')
ax.legend(loc = 10)
```

```python
fig,axes = plt.subplots(nrows = 1,ncols =5,figsize=(8,4))
plt.tight_layout()
for current in axes:
    current.plot(x,y)
    
      
    
```

```python
import numpy as np
x = np.arange(0,100)
y = x*2
z=x**2
```

```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
ax.plot(x,y)
```

```python
fig = plt.figure()
axes1 = fig.add_axes([0,0,1,1])
axes2 = fig.add_axes([.2,.5,.2,.2])
axes1.plot(x,y)
axes1.set_title('bigger plot')
axes2.plot(x,y)
```

```python
fig = plt.figure()
axes1 = fig.add_axes([0,0,1,1])
axes2 = fig.add_axes([.2,.5,.4,.4])
axes1.plot(x,y,z)
axes1.set_title('bigger plot')
axes2.plot(x,y,z)
```
